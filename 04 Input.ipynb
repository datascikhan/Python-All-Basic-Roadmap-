
## 🔸 Typecasting Errors & Handling
📘 Explanation:
- Agar user invalid value input kare jo int() ya float() mein convert nahi ho sakti, toh error (ValueError) aata hai.

🗣️ Roman Urdu:
- Agar user likhe "hello" aur aap int("hello") karen, toh error aayega. Is liye hamesha try aur except ka use karna chahiye.

```python
try:
    age = int(input("Apni age likhein: "))
    print("Aapki age hai:", age)
except ValueError:
    print("Galat input! Sirf numbers likhein.")
```

```python
try:
    weight = float(input("Apna wazan (kg): "))
    print("Wazan:", weight)
except ValueError:
    print("Please decimal ya number likhein.")
```

```python
try:
    num = int(input("Ek number likhein: "))
    print("Square:", num ** 2)
except:
    print("Invalid input!")
```

```python
try:
    price = float(input("Product price likhein: "))
    qty = int(input("Quantity: "))
    print("Total:", price * qty)
except:
    print("Kisi value mein ghalti hai.")
```

```python
try:
    exp = eval(input("Expression likhein: "))
    print("Result:", exp)
except:
    print("Invalid expression!")
```

```python
try:
    a = int(input("Pehla Number likhein."))
    b = int(input("Dusra Number Likhein."))
    print("Division:", a / b)
except ZeroDivisionError:
    print("Zero Se Divide nhi kar sakhte ho.")
except:
    print("invalid input")
```

```python
try:
    marks = int(input("Marks Likhein"))
    if marks < 0:
        raise ValueError("Negative marks nhi ho sakhta")
    print("Marks:", marks)
except Exception as e:
    print("Error:", e)
```

# 🔹 Error Handling

- Programming mein errors (ya exceptions) ka aana normal baat hai. Jab user galat input deta hai, toh program crash na ho, balki use samajhayein ke galat input diya aur dobara try karein. Python mein ye kaam try-except blocks se hota hai. Isse hum error ko pakad ke program ko smoothly chalate hain.

# 🔸 1. Handling invalid input with try-except
Explanation:

- try block mein wo code rakho jahan error aa sakta hai. Agar error aata hai toh program except block mein chala jata hai, jahan aap error ko handle kar sakte ho.

```python
try:
    val = int(input("Number likho"))
    print(f"Value: {val}")

except ValueError:
    print("Sirf Number not Text.")
```

```python
try:
    val = int(input('Number likho'))
    print(f"Only Number: {val}.")

except ValueError:
    print(f"Sirf Number likho not Text.")
```

```python
try:
    age = int(input("Age likho"))
    print(f"Age: {age}.")

except ValueError:
    print("Age Integer Number mein hota hai not Decimal Number mein.")
```

```python
try:
    val = int(input("Integer Number do"))
    print(f"Integer Numbers: {val}.")

except ValueError:
    print("Integer Number do not Decimal Number")
```

```python
try:
    x = int(input("Integer do: "))
except ValueError:
    print("Sirf integer!")
```

```python
try:
    val = int(input("Enter value: "))
except Exception as e:
    print("Error aaya:", e)
```

```python
try:
    score = int(input("Score do: "))
    if score < 0:
        raise ValueError("Score negative nahi ho sakta")
except ValueError as e:
    print("Error:", e)
```

```python
try:
    n = int(input("Number do: "))
    print(10 / n)
except ZeroDivisionError:
    print("Zero se divide nahi kar sakte!")
except ValueError:
    print("Number do!")
```

```python
try:
    val = int(input("Input: "))
    print(val**2)
except:
    print("Galat input hai!")
```

# 🔸 3. Creating reusable input functions
Explanation:

- Input lena aur validate karna baar-baar hota hai, isliye function banakar baar baar reuse kar sakte hain. Is tarah code clean aur maintainable rehta hai.

```python
def get_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("Sirf Integer do")

num = get_int("Number do:")
print(num)
```

```python
def get_float(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Decimal do!")

val = get_float("Decimal do: ")
print(val)
```

```python
def get_positive_int(prompt):
    while True:
        try:
            n = int(input(prompt))
            if n > 0:
                return n
            else:
                print("Positive number do!")
        except:
            print("Sirf number do!")

age = get_positive_int("Age do: ")
print(age)
```

```python
def get_choice(prompt, options):
    while True:
        val = input(prompt).lower()
        if val in options:
            return val
        print("Invalid choice!")

choice = get_choice("Choice (y/n): ", ['y', 'n'])
print(choice)
```

```python
def get_input(prompt, type_):
    while True:
        try:
            return type_(input(prompt))
        except ValueError:
            print(f"Please enter a valid {type_.__name__}")

n = get_input("Number do: ", int)
print(n)
```

```python
def get_nonempty_string(prompt):
    while True:
        s = input(prompt).strip()
        if s:
            return s
        print("Empty input nahi chalega!")

name = get_nonempty_string("Apna naam do: ")
print(name)
""
```

```python
def get_valid_number(prompt):
    while True:
        try:
            val = int(input(prompt))
            return val
        except ValueError:
            print("Number do!")

num = get_valid_number("Number do: ")
print(num)
```

```python
def input_with_validation(prompt, validate):
    while True:
        val = input(prompt)
        if validate(val):
            return val
        print("Invalid input!")

def is_digit(val):
    return val.isdigit()

num = input_with_validation("Number do: ", is_digit)
print(num)
```

## 🔸 4. Unit testing code using input()
Explanation:

- input() wali code ko unit test karne ke liye unittest.mock module mein patch() function use hota hai, jisse hum input ko mock kar sakte hain taake automated testing ho sake.


```python
# 1 - Basic patching input
from unittest.mock import patch

def get_name():
    return input("Name: ")

with patch('builtins.input', return_value='Ali'):
    print(get_name())  # Output: Ali
```

```python
def add_two_numbers():
    a = int(input())
    b = int(input())
    return a + b

with patch('builtins.input', side_effect=['3', '5']):
    print(add_two_numbers())  # Output: 8
```

```python
def get_positive():
    while True:
        n = input()
        if n.isdigit() and int(n) > 0:
            return int(n)

with patch('builtins.input', side_effect=['-1', '0', '4']):
    print(get_positive())  # Output: 4
```

```python
import unittest

class TestInput(unittest.TestCase):
    @patch('builtins.input', return_value='TestUser')
    def test_name(self, mock_input):
        self.assertEqual(get_name(), 'TestUser')

if __name__ == '__main__':
    unittest.main()
```

```python
def multiply_inputs():
    x, y = int(input()), int(input())
    return x * y

with patch('builtins.input', side_effect=['2', '6']):
    print(multiply_inputs())  # Output: 12

```
