## 2. Negative Indexing

- Negative indexing se hum list ke end se elements ko access kar sakte hain.

- Negative indexes -1 se start hote hain jo last element ko point karta hai.

- Negative indexing useful hai jab hume list ke aakhir ke elements chahiye hon.


```python
# Example 1: Last element using -1
fruits = ["apple", "banana", "cherry"]
print("Negative Indexing:", fruits[-1])  # cherry
```

## 3. Slicing

- Slicing se hum list ke ek subset (chhota hissa) ko nikal sakte hain.

- Syntax: list[start:end]  ye start index se lekar end-1 index tak ke elements deta hai.

- start se slice shuru hoti hai, lekin end index tak nahi jati (end exclusive hota hai).

- Agar start ya end specify na karen to woh default se start ya end tak consider hota hai.

- Hum step bhi specify kar sakte hain: list[start:end:step]  jisse hum elements ko step-wise le sakte hain (har second element, third element, etc).


```python
# Example 1: Basic slicing from index 0 to 2 (2 excluded)
fruits = ["apple", "banana", "cherry", "date", "fig"]
print("Slicing:", fruits[0:2])  # ['apple', 'banana']
```

```python
# Example 2: Slicing from index 1 to 4 (4 excluded)
print("Slicing:", fruits[1:4])  # ['banana', 'cherry', 'date']
```

```python
# Example 3: Slicing from start to index 3
print("Slicing:", fruits[:3])   # ['apple', 'banana', 'cherry']
```

```python
# Example 4: Slicing from index 2 till end
print("Slicing:", fruits[2:])   # ['cherry', 'date', 'fig']
```

```python
# Example 5: Slicing whole list (copy)
print("Slicing:", fruits[:])    # ['apple', 'banana', 'cherry', 'date', 'fig']
```

```python
# Example 6: Slicing with step (every 2nd element)
print("Slicing:", fruits[0:5:2])  # ['apple', 'cherry', 'fig']
```

```python
# Example 7: Negative slicing - reverse the list using slicing
print("Slicing:", fruits[::-1])  # ['fig', 'date', 'cherry', 'banana', 'apple']
```

```python
# Example 8: Slicing with negative step from index 3 to 0
print("Slicing:", fruits[3:0:-1])  # ['date', 'cherry', 'banana']
```

# Step 3: Modifying Lists

- Changing Elements

- Adding Elements (append(), insert(), extend())

- Removing Elements (remove(), pop(), clear())

- Python lists ko mutable kehte hain, iska matlab aap list ke items ko change, add, ya remove kar sakte hain. Yeh section teen parts mein hai:

- Changing Elements (items ko update karna)

- Adding Elements (items ko list mein add karna)

- Removing Elements (items ko list se remove karna)

## 1. Changing Elements

- List ke specific element ko update karne ke liye, aap us element ki index ka use karte hain, aur usko nayi value assign kar dete hain.

Syntax:

list_name[index] = new_value


```python
# Example 1: Change first element
fruits = ["apple", "banana", "cherry"]
print("Before:", fruits)
fruits[0] = "mango"  # Change Element (apple)
print("After:", fruits)
```

```python
# Example 2: Change last element
numbers = [10, 20, 30, 40]
print("Before:", numbers)
numbers[-1] = 50
print("After:", numbers)
```

```python

# Example 3: Change element in nested list
nested = [1, 2, [10, 20, 30]]
print("Before:", nested)
nested[2][1] = 99
print("After:", nested)
```

```python
# Example 4: Change multiple elements using slicing
letters = ['a', 'b', 'c', 'd', 'e']
print("Before:", letters)
letters[1:3] = ['x', 'y']
print("After:", letters)
```

```python
# Example 5: Replace single element with multiple elements via slicing
letters = ['a', 'b', 'c', 'd', 'e']
print("Before:", letters)
letters[2:3] = ['x', 'y', 'z']
print("After:", letters)
```

## 2. Adding Elements (append(), insert(), extend())

- append() method list ke end mein ek element add karta hai.

- insert(index, element) method kisi specific position (index) pe element insert karta hai, baaki elements shift ho jate hain.

- extend(iterable) method doosri list ya iterable ke saare elements ko current list ke end mein add karta hai.


```python
# Example 1: append() adds element at the end
fruits = ["apple", "banana"]
print("Before append:", fruits)
fruits.append("cherry")
print("After append:", fruits)
```

```python
# Example 2: append() with integer list
numbers = [1, 2, 3]
print("Before append:", numbers)
numbers.append(4)
print("After append:", numbers)
```

```python
# Example 3: insert() element at index 1
fruits = ["apple", "banana", "cherry"]
print("Before insert:", fruits)
fruits.insert(1, "mango")
print("After insert:", fruits)
```

```python
# Example 4: insert() at beginning
letters = ['b', 'c', 'd']
print("Before insert:", letters)
letters.insert(0, 'a')
print("After insert:", letters)
```

- Membership operators ka use list mein kisi element ki mojoodgi (presence) check karne ke liye hota hai.

# Step 9: Copying Lists

- Shallow Copy (copy(), slicing)

- Deep Copy (copy module)

- Copying Lists kya hai?

- Kabhi kabhi hume ek list ki copy banani padti hai taa ke asli list ko affect kiye bina naye list pe kaam kar saken. Python mein lists ko copy karne ke do main tareeke hain: Shallow Copy aur Deep Copy.

## 1. Shallow Copy (copy(), slicing)

- Shallow copy original list ki ek nayi list object banata hai, lekin agar list ke andar nested (andar ke) mutable objects hain, to woh original aur copy dono mein shared rehte hain.

- Simple baat: sirf first-level objects ka duplicate banta hai, nested objects reference rahte hain.

- Shallow copy karne ke liye hum do tareeke use karte hain:

- list.copy() method

- Slicing [:]

```python

# Example 1: Using copy() to shallow copy a list
import copy
original = [1, 2, 3]
shallow = original.copy()
print("Original:", original)
print("Shallow Copy:", shallow)
```

## 1. Passing Lists to Functions

- Jab hum list ko function ko argument ke tor pe dete hain, to function ko us list ka reference milta hai, copy nahi.

- Iska matlab function list ko directly modify kar sakta hai.

- Agar function mein list ko change karenge to original list bhi change hogi.

- Function ke andar list ko read karna ya iterate karna bhi ho sakta hai bina change kiye.


```python
# Example 1: Simple function jo list print karta hai
def print_list(lst):
    for item in lst:
        print(item)

fruits = ["apple", "banana", "cherry"]
print_list(fruits)
```

```python

# Example 2: Function jo list elements ko double karta hai (modify karta hai)
def double_list(lst):
    for i in range(len(lst)):
        lst[i] = lst[i] * 2

numbers = [1, 2, 3]
double_list(numbers)
print("Doubled:", numbers)
```

```python
# Example 3: Function jo list ke elements ka sum print karta hai
def sum_list(lst):
    total = 0
    for num in lst:
        total += num
    print("Sum:", total)

nums = [10, 20, 30]
sum_list(nums)
```

```python
# Example 4: Function jo list mein naye elements add karta hai
def add_elements(lst):
    lst.append(100)
    lst.append(200)

my_list = [1, 2, 3]
add_elements(my_list)
print("After adding:", my_list)
```

```python
# Example 5: Function jo list mein se negative numbers remove karta hai
def remove_negatives(lst):
    i = 0
    while i < len(lst):
        if lst[i] < 0:
            lst.pop(i)
        else:
            i += 1

nums = [5, -1, 3, -2, 0]
remove_negatives(nums)
print("No negatives:", nums)
```

```python
# Example 6: Function jo list ko iterate karke elements print karta hai
def iterate_list(lst):
    for element in lst:
        print("Element:", element)

colors = ["red", "green", "blue"]
iterate_list(colors)
```

```python
# Example 7: Function jo list ke elements ko square karta hai (modifies original)
def square_elements(lst):
    for i in range(len(lst)):
        lst[i] = lst[i] ** 2

nums = [1, 2, 3]
square_elements(nums)
print("Squares:", nums)
```

```python
# Example 8: Function jo list ke har element ke sath ek string concatenate karta hai
def add_suffix(lst, suffix):
    for i in range(len(lst)):
        lst[i] = lst[i] + suffix

names = ["Ali", "Sara", "Ahmed"]
add_suffix(names, "_2025")
print("Modified names:", names)
```

## 2. Returning Lists from Functions

- Functions lists ko return kar sakte hain.

- Kabhi kabhi hum nayi list banake function se return karte hain taa ke original list untouched rahe.

- Function ke andar hum list ko create karke ya modify karke return kar sakte hain.

- Returned list ko hum variable mein store kar sakte hain aur use kar sakte hain.


```python
# Example 1: Function jo list return karta hai
def get_fruits():
    fruits = ["apple", "banana", "cherry"]
    return fruits

my_fruits = get_fruits()
print(my_fruits)
```

```python
# Example 2: Function jo numbers ka square list return karta hai
def squares(nums):
    result = []
    for n in nums:
        result.append(n ** 2)
    return result

nums = [1, 2, 3]
print(squares(nums))
```

```python
# Example 3: Function jo list ke positive numbers return karta hai
def positive_numbers(lst):
    positives = []
    for num in lst:
        if num > 0:
            positives.append(num)
    return positives

numbers = [-1, 3, -5, 7, 0]
print(positive_numbers(numbers))
```

```python
# Example 4: Function jo list reverse karke return karta hai (original ko change nahi karta)
def reverse_list(lst):
    reversed_list = lst[::-1]
    return reversed_list

data = [10, 20, 30]
print(reverse_list(data))
print("Original:", data)
```

```python
# Example 5: Function jo string list ke lengths ki list return karta hai
def lengths(lst):
    lengths_list = []
    for s in lst:
        lengths_list.append(len(s))
    return lengths_list

words = ["hello", "world", "python"]
print(lengths(words))
```

```python
# Example 7: Function jo list ke elements ko uppercase karke return karta hai
def to_upper(lst):
    result = []
    for item in lst:
        result.append(item.upper())
    return result

names = ["ali", "sara", "ahmed"]
print(to_upper(names))
```

```python
# Example 8: Function jo list ke unique elements return karta hai
def unique_elements(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

items = [1, 2, 2, 3, 4, 4, 5]
print(unique_elements(items))
```

## 1. Mutability Concept

- Mutability ka matlab hai ke ek object ke contents ko modify kiya ja sakta hai bina us object ka naya object banaye.

- Python mein lists mutable hoti hain, iska matlab hum list ke elements ko badal sakte hain, add kar sakte hain, ya remove kar sakte hain bina nayi list banaye.

- Mutable objects ka behaviour reference ke sath hota hai  jab hum list ko kisi variable mein assign karte hain, wo reference hota hai asli list object ka.


```python
# Example 1: List elements ko change karna (mutability)
my_list = [1, 2, 3]
my_list[0] = 100
print("Modified list:", my_list)  # Output: [100, 2, 3]
```

```python
# Example 2: List mein naye element add karna (mutability)
fruits = ["apple", "banana"]
fruits.append("cherry")
print("Fruits after append:", fruits)
```

```python
# Example 3: List mein element remove karna (mutability)
nums = [10, 20, 30, 40]
nums.remove(20)
print("After removal:", nums)
```

```python
# Example 4: List ke multiple elements ko update karna
values = [1, 2, 3, 4, 5]
values[1:4] = [20, 30, 40]
print("Updated values:", values)  # Output: [1, 20, 30, 40, 5]
```

```python
# Example 5: Mutability ka effect function mein
def modify_list(lst):
    lst.append(99)

numbers = [1, 2, 3]
modify_list(numbers)
print("After function call:", numbers)  # Output: [1, 2, 3, 99]
```

```python
# Example 6: Mutability ke bawajood original list ka reference same rehta hai
lst = [1, 2, 3]
print("Before modification:", id(lst))
lst.append(4)
print("After modification:", id(lst))  # id same rahega
```

## 2. References vs Copies

- Jab hum ek list ko doosre variable mein assign karte hain, wo reference copy hota hai, yani dono variables same list ko point karte hain.

- Agar hum copy banate hain (shallow ya deep copy), to naya alag object banta hai.

- References se change original list mein reflect hota hai, copies mein nahi.

- Ye concept bohat important hai taake bugs na hon jab hum lists functions mein ya multiple variables mein use kar rahe hon.

```python
# Example 1: Reference assignment
a = [1, 2, 3]
b = a  # b ab a ka reference hai
b.append(4)
print("a:", a)  # Output: [1, 2, 3, 4]
print("b:", b)  # Output: [1, 2, 3, 4]
```

```python
# Example 2: Shallow copy using slicing
a = [10, 20, 30]
b = a[:]  # Shallow copy
b.append(40)
print("a:", a)  # Output: [10, 20, 30]
print("b:", b)  # Output: [10, 20, 30, 40]
```

```python
# Example 3: Shallow copy and nested list issue
a = [[1, 2], [3, 4]]
b = a[:]  # Shallow copy
b[0][0] = 999
print("a after change:", a)  # Nested element changed!
print("b:", b)
```

```python
# Example 4: Deep copy to avoid nested list mutation
import copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
b[0][0] = 999
print("a:", a)  # Original untouched
print("b:", b)  # Modified deep copy
```

```python
# Example 5: Reference vs copy id comparison
a = [1, 2, 3]
b = a
c = a[:]
print("id(a):", id(a))
print("id(b):", id(b))  # Same as a
print("id(c):", id(c))  # Different from a
```

```python

# Example 6: Function example with reference
def modify(lst):
    lst.append(100)

lst1 = [1, 2]
modify(lst1)
print(lst1)  # [1, 2, 100]
```

```python
# Example 7: Function example with copy
def modify(lst):
    lst = lst[:]  # Local copy banai
    lst.append(200)
    print("Inside function:", lst)

lst2 = [1, 2]
modify(lst2)
print("Outside function:", lst2)  # Original unchanged
```

```python
# Example 8: Using copy module for shallow copy
import copy
original = [1, 2, 3]
shallow_copy = copy.copy(original)
shallow_copy.append(4)
print("Original:", original)
print("Shallow Copy:", shallow_copy)
```

##  3. Using filter() and map() with Lists

- filter() use hota hai list se specific elements nikalne ke liye (based on condition).

- Dono functions first-class functions hain jo lambda ya normal function accept karte hain.

- tuple  Immutable list
