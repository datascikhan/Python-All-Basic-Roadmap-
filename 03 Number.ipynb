# 3 Complex Numbers (complex)
 - Complex numbers real aur imaginary dono parts rakhte hain, jese 2 + 3j, jahan j imaginary unit hai.
 - A complex number has a real part and an imaginary part.
 - a = 2 + 3j
 - 2 is the real part
 - 3j is the imaginary part

```python
c1 = 2 + 3j
c2 = complex(5, -1)
print(c1, c2)               # (2+3j) (5-1j)
print(type(c1))             # <class 'complex'>
```

```python
# Real and imaginary parts
print("Real Part:",c1.real)              # 2.0
print("Imaginary Part:",c1.imag)              # 3.0
```

```python
# Complex arithmetic
print(c1 + c2)              # (7+2j)
print(c1 * c2)              # (13+13j)
print(abs(c1))              # Magnitude sqrt(2^2 + 3^2) = 3.605551275463989
print(c1.conjugate())       # (2-3j)
```

```python
z = 4 + 5j   # Complex Number
print("Complex number:", z)
```

### 1 Addition (+), Subtraction (-), Multiplication (*), Division (/)
 - Addition (+): Do numbers ko jorna
 - Subtraction (-): Do numbers mein se aik ko minus karna

```python
a = 10
b = 3

print("Addition",a + b)    # 13
print("Subtraction",a - b)    # 7
print("Multiplication",a * b)    # 30
print("Division",a / b)    # 3.3333333333333335
print(15 + 5)   # 20
print(100 - 45) # 55
```

### 2 Floor Division (//)

- Floor division se division ka sirf integer part milta hai, decimal part chhor diya jata hai.


```python

a = 10
b = 3

print("Floor Division:",a // b)   # 3  (10/3 = 3.33, floor division = 3)
print("Floor Division:",15 // 4)  # 3
print("Floor Division:",20 // 7)  # 2
print("Floor Division:",-10 // 3) # -4 (floor division rounds down)
print("Floor Division:",9 // 3)   # 3
```

### 3 Modulus (%)

```python
a = 10
b = 3

print("Remainder:",a % b)    # 1  (10 divided by 3 remainder 1)
print("Remainder:",15 % 4)   # 3
print("Remainder:",20 % 7)   # 6
print("Remainder:",9 % 3)    # 0
print("Remainder:",7 % 5)    # 2
```

# Step 4: Built-in Numeric Functions

### 1 abs()  Absolute value
 - abs() function kisi bhi number ka absolute value (magnitude) deta hai, yaani negative number ko positive bana deta hai.


 - Agar number positive ya zero hai, toh wahi return hota hai.

 - Agar number negative hai, toh uska positive equivalent return hota hai.

```python
print(abs(10))      # 10
print(abs(-5))      # 5
print(abs(0))       # 0
print(abs(-3.14))   # 3.14
print(abs(complex(3, -4)))  # 5.0 (complex number ka magnitude)
```

### 2 round()  Number ko round karna
 - round() function decimal numbers ko nearest integer ya specified decimal places tak round karta hai.


 - Agar sirf number diya jaye toh nearest integer round hota hai.

 - Agar dusra argument diya jaye, toh utne decimal places tak round karega.

```python
print(round(3.14159))         # 3
print(round(3.14159, 2))      # 3.14
print(round(2.71828, 3))      # 2.718
print(round(5.5))             # 6   (Python 3 mein half to even rounding)
print(round(-1.2345, 2))      # -1.23
```

### 3 pow()  Power calculate karna

 - pow(base, exponent) function base ka exponent power nikalta hai.

 - pow(a, b) matlab a ki power b.

 - pow(a, b, mod) bhi hota hai jo modular exponentiation karta hai.

```python
print(pow(2, 3))          # 8   (2^3)
print(pow(5, 0))          # 1   (koi bhi number 0 power 1 hota hai)
print(pow(10, 2))         # 100
print(pow(3, 4, 5))       # 1   (3^4 % 5 = 81 % 5 = 1)
print(pow(2, -1))         # 0.5 (power negative bhi ho sakta hai)
```

### 4 divmod()  Division aur modulus ek saath return karna
 - divmod(a, b) function do cheezein return karta hai: (quotient, remainder) yaani division ka result aur remainder ek tuple mein.

 - divmod(10, 3) return karega (3, 1)

 - Useful hai jab aapko integer division aur remainder dono chahiye.

```python
print(divmod(10, 3))      # (3, 1)
print(divmod(15, 4))      # (3, 3)
print(divmod(20, 7))      # (2, 6)
print(divmod(9, 3))       # (3, 0)
print(divmod(-10, 3))     # (-4, 2) (floor division ka rule follow karta hai)
```

### Step 6: Random Number Generation

 - import random

 - random.random()  0 se 1 ke beech random float

 - random.randint(a, b)  a se b ke beech random integer

 - Python ka random module random numbers aur random selections generate karne ke liye use hota hai. Pehle import random karna zaroori hota hai.



### 1 random.random()

 - 0 se 1 ke beech (inclusive 0, exclusive 1) random float number generate karta hai.

 - Har call par ek random decimal number return karta hai jo 0.0 se 1.0 ke beech hota hai.

 - Kabhi bhi exactly 1.0 nahi hota.


```python
import random

# 1
print(random.random())       # e.g., 0.37444887175646646

# 2 - 5 random numbers ek list mein
nums = [random.random() for _ in range(5)]
print(nums)                  # e.g., [0.56, 0.91, 0.12, 0.67, 0.23]

# 3 - Random number between 0 and 1 with rounding
print(round(random.random(), 2))  # e.g., 0.73

# 4 - Use random.random() to simulate probability (50% chance)
if random.random() < 0.5:
    print("Heads")
else:
    print("Tails")

# 5 - Generate random float and scale it to 0 to 10
scaled = random.random() * 10
print(scaled)               # e.g., 7.435

# 6 - Generate random float in custom range (e.g., 5.5 to 9.5)
def random_float_range(start, end):
    return start + (end - start) * random.random()

print(random_float_range(5.5, 9.5))  # e.g., 7.832
```


### 2 random.randint(a, b)

 - a se b ke beech inclusive random integer generate karta hai.

 - Dono a aur b shamil (inclusive) hote hain.

 - Integer random value return karta hai.

```python
import random

# 1 - Random integer between 1 and 10
print(random.randint(1, 10))     # e.g., 7

# 2 - Random integer between -5 and 5
print(random.randint(-5, 5))     # e.g., 0

# 3 - Random integer between 100 and 200
print(random.randint(100, 200))  # e.g., 153

# 4 - Using randint in dice roll simulation (1 to 6)
dice = random.randint(1, 6)
print("Dice roll:", dice)        # e.g., 3

# 5 - Generating 5 random integers between 10 and 20
rand_nums = [random.randint(10, 20) for _ in range(5)]
print(rand_nums)                 # e.g., [11, 14, 17, 20, 12]

# 6 - Random number with same start and end (always same)
print(random.randint(5, 5))      # 5
```

### 3 random.choice()

 - List ya sequence mein se random element select karta hai.

 - Kisi bhi list, tuple, string se ek random element return karta hai.

 - Input iterable hona chahiye.

```python
import random

fruits = ['apple', 'banana', 'mango', 'orange', 'grapes']

# 1 - Random fruit select karo
print(random.choice(fruits))     # e.g., mango

# 2 - Random character from string
print(random.choice("hello"))    # e.g., 'e'

# 3 - Random element from numbers list
numbers = [10, 20, 30, 40, 50]
print(random.choice(numbers))    # e.g., 30

# 4 - Random element from tuple
colors = ('red', 'green', 'blue')
print(random.choice(colors))     # e.g., green

# 5 - Use random.choice in a function to pick random greeting
def random_greeting():
    greetings = ["Hello", "Hi", "Salam", "Namaste", "Hey"]
    return random.choice(greetings)

print(random_greeting())          # e.g., Salam

# 6 - Random selection from list multiple times
for _ in range(3):
    print(random.choice(fruits))
# e.g., apple, orange, grapes
```

# Step 8: Bitwise Operators (Numbers par operations)
Explanation:

-Bitwise operators numbers ke binary representation par kaam karte hain. Matlab yeh operators har bit (0 ya 1) ke saath individually operation karte hain.

-Python mein commonly yeh operators use hote hain:

 - AND (&)

 - OR (|)

 - XOR (^)

 - NOT (~)

 - Left Shift (<<)

 - Right Shift (>>)


# 1 AND (&)

- Har bit ke liye:

- 1 & 1 = 1

 - Baaki sab combinations = 0
 - Matlab dono bits 1 hon tabhi result 1 hoga.

```python
a = 10          # binary: 1010
b = 6           # binary: 0110

# 1
print(a & b)    # 2 (0010)

# 2
print(12 & 5)   # 4  (1100 & 0101 = 0100)

# 3
print(7 & 3)    # 3  (0111 & 0011 = 0011)

# 4
print(0 & 15)   # 0  (0000 & 1111 = 0000)

# 5
print(255 & 128) # 128 (11111111 & 10000000 = 10000000)

# 6 - Bitwise AND with same number
print(42 & 42)  # 42

# 7 - AND with zero
print(42 & 0)   # 0

# 8 - AND with all ones (for 8-bit)
print(42 & 0xFF)  # 42
```

### 2 OR (|)

- Har bit ke liye:

- Agar dono mein se koi bhi bit 1 ho toh result 1 hoga.


```python
a = 10          # binary: 1010
b = 6           # binary: 0110

# 1
print(a | b)    # 14 (1110)

# 2
print(12 | 5)   # 13 (1100 | 0101 = 1101)

# 3
print(7 | 3)    # 7  (0111 | 0011 = 0111)

# 4
print(0 | 15)   # 15 (0000 | 1111 = 1111)

# 5
print(255 | 128) # 255 (11111111 | 10000000 = 11111111)

# 6 - OR with same number
print(42 | 42)  # 42

# 7 - OR with zero
print(42 | 0)   # 42

# 8 - OR with all zeros
print(0 | 0)    # 0
```

### 3 XOR (^)

- Har bit ke liye:

- Agar dono bits alag-alag hon toh result 1 hoga, warna 0.

### 5 Left Shift (<<)

- Binary number ko left shift karta hai, yaani bits ko left move karta hai aur right mein zeros add karta hai.

- x << n matlab x ko n bits left shift karo  number x * (2**n) ke barabar ho jata hai.


```python
# 1 - 2 ko 1 bit left shift karo (2 * 2 = 4)
print(2 << 1)    # 4

# 2 - 3 ko 2 bit left shift karo (3 * 4 = 12)
print(3 << 2)    # 12

# 3 - 5 ko 3 bit left shift karo (5 * 8 = 40)
print(5 << 3)    # 40

# 4 - 1 ko 4 bit left shift karo (1 * 16 = 16)
print(1 << 4)    # 16

# 5 - 10 ko 0 bit left shift karo (same number)
print(10 << 0)   # 10

# 6 - 7 ko 5 bit left shift karo (7 * 32 = 224)
print(7 << 5)    # 224

# 7 - 0 ko kisi bhi shift se kuch nahi hota
print(0 << 10)   # 0

# 8 - 8 ko 1 bit left shift karo (8 * 2 = 16)
print(8 << 1)    # 16
```

### 6 Right Shift (>>)

- Binary number ko right shift karta hai, yaani bits ko right move karta hai aur left mein zeros add karta hai (positive numbers ke liye).

- x >> n matlab x ko n bits right shift karo  number x // (2**n) ke barabar ho jata hai.


```python
# 1 - 4 ko 1 bit right shift karo (4 // 2 = 2)
print(4 >> 1)    # 2

# 2 - 12 ko 2 bit right shift karo (12 // 4 = 3)
print(12 >> 2)   # 3

# 3 - 40 ko 3 bit right shift karo (40 // 8 = 5)
print(40 >> 3)   # 5

# 4 - 16 ko 4 bit right shift karo (16 // 16 = 1)
print(16 >> 4)   # 1

# 5 - 10 ko 0 bit right shift karo (same number)
print(10 >> 0)   # 10

# 6 - 224 ko 5 bit right shift karo (224 // 32 = 7)
print(224 >> 5)  # 7

# 7 - 0 ko kisi bhi shift se kuch nahi hota
print(0 >> 10)   # 0

# 8 - 16 ko 1 bit right shift karo (16 // 2 = 8)
print(16 >> 1)   # 8
```

###  2. Prime Number Checking Logic
 Explanation:

- Prime number wo hota hai jo:

- Greater than 1 ho

- Sirf 1 aur khud se divide ho sakta ho

 Logic:

- 1 se chhoti values prime nahi hoti
- Check karna hota hai ke kisi number se divide hota hai ya nahi from 2 to sqrt(n)


```python
# 1 - Basic prime check
num = 7
is_prime = True
if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

print(f"{num} is prime? {is_prime}")
```

```python
# 2 - Function form
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(11))  # True
```

```python
# 3 - Prime check with 1 and 0
print(is_prime(1))   # False
print(is_prime(0))   # False

# 4 - Negative number
print(is_prime(-7))  # False
```

```python
# 5 - List of prime numbers from 1 to 20
for n in range(1, 21):
    if is_prime(n):
        print(n, "is prime")
```

```python
# 6 - Early return optimization
def fast_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(fast_prime(29))  # True
```

```python
# 7 - Prime check inside input
x = int(input("Enter number: "))
print("Prime" if is_prime(x) else "Not Prime")
```
